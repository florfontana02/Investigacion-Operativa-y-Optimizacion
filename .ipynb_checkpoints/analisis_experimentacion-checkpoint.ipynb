```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.transforms import Affine2D
```

```python
ejemplos = []
ej1 = pd.read_csv("resultados_parametros_ej1.csv")
ejemplos.append(["ej1",ej1])
ej2 = pd.read_csv("resultados_parametros_ej2.csv")
ejemplos.append(["ej2",ej2])
ej3 = pd.read_csv("resultados_parametros_ej3.csv")
ejemplos.append(["ej3",ej3])
ej4 = pd.read_csv("resultados_parametros_ej4.csv")
ejemplos.append(["ej4",ej4])
ej5 = pd.read_csv("resultados_parametros_ej5.csv")
ejemplos.append(["ej5",ej5])
```

```python
palette = sns.color_palette("viridis")
def grafico_barras(nombre,ej):
    #sns.set(style="whitegrid")
    g = sns.catplot(data=ej,x='Parámetro',y='Tiempo (s)',hue='Valor',kind='bar', palette=palette)

    g.set_axis_labels("Parámetro","Tiempo (s)")
    plt.xticks(rotation=45)

    plt.title(f"{nombre}")
    dx, dy = -50, 0  # Ajusta estos valores según sea necesario
    offset = Affine2D().translate(dx, dy)
    for label in g.ax.get_xticklabels():
        label.set_transform(label.get_transform() + offset)

    plt.show()
```

```python
for ejemplo in ejemplos:
    grafico_barras(ejemplo[0],ejemplo[1])
```

```python
def promedios_tiempos(nombre,ej):
    categorias= ej["Parámetro"].unique()
    res = []
    for cat in categorias:
        valores = ej[ej["Parámetro"]==cat]["Tiempo (s)"]
        promedio = valores.describe()['mean']
        res.append(promedio)
    
    # Grafico
    plt.bar(categorias,res)
    plt.title(nombre)
    plt.xlabel("Categorias")
    plt.ylabel("Promedio Tiempo (s)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
for ejemplo in ejemplos:
    promedios_tiempos(ejemplo[0],ejemplo[1])
```

```python
def promedio_categorias(ejemplos):
    categorias = ejemplos[0][1]["Parámetro"].unique()
    res = []
    for cat in categorias:
        prom = 0
        cant = 0
        for nombre,ej in ejemplos:
            lista = ej[ej["Parámetro"]==cat]["Tiempo (s)"]
            cant += len(lista)
            prom += lista.sum()
        res.append(prom/cant)
    
    plt.bar(categorias,res)
    plt.title("Promedio de todos los ejemplos por categoria")
    plt.xlabel("Categorias")
    plt.ylabel("Promedio Tiempo (s)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
promedio_categorias(ejemplos)
```

```python
ejemplos = []
for i in range(1, 6):
    ej = pd.read_csv(f"resultados_parametros_ej{i}.csv")
    ejemplos.append(ej)

    
#Me quedo con las columnas de Parámetro, Valor, Tiempo y Objetivo    
filas_extraidas = [ej.iloc[:, :4] for ej in ejemplos]

#Promedio las columnas 3 y 4
df_prom = filas_extraidas[0][['Parámetro', 'Valor']].copy() 
df_prom['Tiempo (s)'] = sum(ej['Tiempo (s)'] for ej in filas_extraidas) / len(filas_extraidas)  
df_prom['Objetivo'] = sum(ej['Objetivo'] for ej in filas_extraidas) / len(filas_extraidas)  


print(df_prom)
```

```python
grafico_barras("Tiempo de Ejecución por Parámetro y Valor", df_prom)
```

```python
df_prom.columns = df_prom.columns.str.strip()
df_prom['Tiempo (s)'] = pd.to_numeric(df_prom['Tiempo (s)'], errors='coerce')


#Creo gráficos de barras para cada categoría de 'Parámetro'
parametros = df_prom['Parámetro'].unique()

for parametro in parametros:
    df_param = df_prom[df_prom['Parámetro'] == parametro]
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_param, x='Valor', y='Tiempo (s)', palette = palette)
    plt.title(f'Tiempo de Ejecución para {parametro}')
    plt.xlabel('Valor')
    plt.ylabel('Tiempo (s)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
